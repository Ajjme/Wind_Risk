```python
import pandas as pd
data = {
    "RCP": ["RCP 2.6", "RCP 4.5", "RCP 8.5"],
    ".1": [115, 118, 122],  # Wind speeds in mph
    ".05": [120, 122, 128],
    ".02": [125, 128, 135],
    ".01": [130, 135, 144],
    ".002": [138, 144, 150],
}

# Create the DataFrame
wind_speed_df = pd.DataFrame(data)

# Display the DataFrame
wind_speed_df
```

```python
import matplotlib.pyplot as plt

# Transpose the dataframe for plotting
transposed_df = wind_speed_df.set_index('RCP').T

# Plot the data
plt.figure(figsize=(10, 6))

# Rename the columns to convert them from strings to numeric values
wind_speed_df.columns = ['RCP'] + [float(col) for col in wind_speed_df.columns[1:]]
# Plot each RCP scenario
for rcp in transposed_df.columns:
    plt.plot(transposed_df.index, transposed_df[rcp], marker='o', label=f"{rcp}")

    
# Adding title and labels
plt.title("Wind Speeds for Different Return Periods Across RCP Scenarios", fontsize=14)
plt.xlabel("Return Period", fontsize=12)
plt.ylabel("Wind Speed (mph)", fontsize=12)

# Adding a legend
plt.legend(title="RCP Scenario")

# Show the plot
plt.grid(True)
plt.show()
```
